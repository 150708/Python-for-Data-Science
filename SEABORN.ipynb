**SEABORN**
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.It's particularly useful for creating complex visualizations such as statistical models, heatmaps, time series, and categorical plots with ease.

```python
#Scatter Plot with Regression Line
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd
import numpy as np

np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)

data = pd.DataFrame({'X': x, 'Y': y})

# Scatter plot with regression line
sns.lmplot(x='X', y='Y', data=data, height=6)
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Histogram
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import numpy as np

np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Histogram with KDE
sns.histplot(data, kde=True)
plt.title('Histogram with KDE')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box Plot

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd
import numpy as np

np.random.seed(0)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value': np.random.normal(size=100)
})

# Box plot
sns.boxplot(x='Category', y='Value', data=data)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import numpy as np

np.random.seed(0)
data = np.random.rand(10, 12)

# Heatmap
sns.heatmap(data, annot=True, cmap='YlGnBu')
plt.title('Heatmap')
plt.show()
```

```python
#Pair Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd
import numpy as np

np.random.seed(0)
data = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])

# Pair plot
sns.pairplot(data)
plt.suptitle('Pair Plot of Variables', y=1.02)
plt.show()
```

```python
#Count Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd

data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'B', 'C', 'C'],
})

# Count plot
sns.countplot(x='Category', data=data)
plt.title('Count Plot')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#Bar Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
import pandas as pd
import numpy as np

np.random.seed(0)
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': np.random.randint(1, 10, size=3)
})

# Bar plot
sns.barplot(x='Category', y='Value', data=data)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```
